```python
import matplotlib.pyplot as plt

from load_data import load_data
from moving_average import moving_average
import datetime
```

```python
markersize = .1
N = 365


header, dates, data = load_data()
```

```python
avgT, maxT, minT = data[:,0], data[:,1], data[:,2]
```

```python
dates_new = dates[dates>datetime.datetime(2013,1,1)]
data_new = data[dates>datetime.datetime(2013,1,1)]
avgT_new, maxT_new, minT_new = data_new[:,0], data_new[:,1], data_new[:,2]
```

```python
plt.plot(dates_new, maxT_new, 'r.', markersize=markersize)
plt.plot(dates_new, minT_new, 'b.', markersize=markersize)
plt.plot(dates_new, avgT_new, 'g.', markersize=2*markersize)

plt.plot(dates_new, moving_average(maxT_new, N), 'r')
plt.plot(dates_new, moving_average(minT_new, N), 'b.')
plt.plot(dates_new, moving_average(avgT_new, N), 'g.')

plt.show()
```
